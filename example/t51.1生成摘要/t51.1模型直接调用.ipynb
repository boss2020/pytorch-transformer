```python
import torch
from pythonicforbert import FullTokenizer
from pythonicforbert import T5Config

from pythonicforbert import get_model_function
import json
MT5ForGeneration,MT5Config,get_data = get_model_function('mt5-base')
import json
with open('/home/xiaoguzai/模型/t5-copy/config.json','r',encoding='utf8')as fp:
    json_data = json.load(fp)
config = MT5Config(**json_data)
model = MT5ForGeneration(config)
model = get_data(model,'/home/xiaoguzai/模型/t5-copy/pytorch_model.bin')
model.eval()
input_ids = torch.tensor([[1,2,3,4,5]])
labels = torch.tensor([[0]])
result,past_key,past_value = model(input_ids,labels)
print('result = ')
print(result)
print('past_key = ')
print(past_key)
print('past_value = ')
print(past_value)
```
